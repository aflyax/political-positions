```python
import numpy as np
from numpy.random import rand
import pandas as pd

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(context="poster", style="dark")
```

```python
votes_113 = pd.read_csv('113_frame.csv')
votes_113.drop('Unnamed: 0', 1, inplace=True)
votes_113.replace('NaN', 2, inplace=True)
votes_113.tail()
```

```python
names = votes_113.name.values
party = votes_113.party.values
state = votes_113.state.values
votes = votes_113[votes_113.columns[0:-3]].values
```

```python
from sklearn.decomposition import PCA
from sklearn.lda import LDA

pca = PCA(n_components=2)
pca_comps = pca.fit(votes).transform(votes)

lda = LDA(n_components=2)
lda_comps = lda.fit(votes, party).transform(votes)

np.shape(pca_comps)
np.shape(lda_comps)
```

```python
colors = votes_113.party.replace({'D': 'b', 'R': 'r', 'I': 'g'})
np.unique(colors)
```

```python
p = list(zip(*pca_comps))
plt.plot(p[0], p[1], '.')
```

```python
for i in range(len(pca_comps)):
    plt.scatter(pca_comps[i,0], pca_comps[i,1], c=colors[i], label=party[i])
```

```python
lda = LDA(n_components=2)
lda_comps = lda.fit(votes, party).transform(votes)

for i in range(len(lda_comps)):
    plt.scatter(lda_comps[i,0], lda_comps[i,1], c=colors[i], label=party[i])
    
# lda takes a greater effort to separate by target categories; e.g., parties
```

```python
# Define some CSS to control our custom labels
css = """
table
{
  border-collapse: collapse;
}
th
{
  color: #ffffff;
  background-color: #000000;
}
td
{
  background-color: #cccccc;
}
table, th, td
{
  font-family:Arial, Helvetica, sans-serif;
  border: 1px solid black;
  text-align: right;
}
"""
```

```python
votes_pca = pd.DataFrame(pca_comps, columns=['pca_1', 'pca_2'])
votes_pca['name'] = names
votes_pca['state'] = state
votes_pca['party'] = party

votes_pca.head()
```

```python
import mpld3
from mpld3 import plugins

fig, ax = plt.subplots()
ax.grid(True, alpha=0.0)

points = ax.scatter(p[0], p[1], c=colors, s=20, alpha=0.8)

labels = []
for i in range(len(votes)):
    label = votes_pca[votes_pca.columns[2:4]].ix[[i], :].T
    label.columns = ['Row {0}'.format(i)]
    # .to_html() is unicode; so make leading 'u' go away with str()
    labels.append(str(label.to_html()))
    
tooltip = plugins.PointHTMLTooltip(points, labels, voffset=10, hoffset=10, css=css)
plugins.connect(fig, tooltip)
plt.title("clustering of 113th US Congress (hover/scroll to interact/see details... tools on lower left)")
mpld3.display()
```

```python
from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3)
pca_3d_comps = pca_3d.fit(votes).transform(votes)

fig = plt.figure()
ax = Axes3D(fig)

p_3 = list(zip(*pca_3d_comps))

ax.scatter3D(p_3[0], p_3[1], p_3[2], c=colors)

# np.shape(p_3)
```

```python
new_votes = votes_113.drop(votes_113.index[[23, 97, 100, 101]], inplace=False)
new_votes.reset_index(inplace=True, drop=True)

names = new_votes.name.values
party = new_votes.party.values
state = new_votes.state.values
votes = new_votes[new_votes.columns[0:-3]].values

pca = PCA(n_components=2)
pca_comps = pca.fit(votes).transform(votes)

p = list(zip(*pca_comps))
# plt.plot(p[0], p[1], '.')

votes_pca = pd.DataFrame(pca_comps, columns=['pca_1', 'pca_2'])
votes_pca['name'] = names
votes_pca['state'] = state
votes_pca['party'] = party

# votes_pca.head()

colors = new_votes.party.replace({'D': 'b', 'R': 'r', 'I': 'g'})

import mpld3
from mpld3 import plugins

fig, ax = plt.subplots()
ax.grid(True, alpha=0.0)

points = ax.scatter(p[0], p[1], c=colors, s=20, alpha=0.8)

labels = []
for i in range(len(votes)):
    label = votes_pca[votes_pca.columns[2:5]].ix[[i], :].T
    label.columns = ['Row {0}'.format(i)]
    # .to_html() is unicode; so make leading 'u' go away with str()
    labels.append(str(label.to_html()))

tooltip = plugins.PointHTMLTooltip(points, labels, voffset=10, hoffset=10, css=css)
plugins.connect(fig, tooltip)
plt.title("clustering of 113th US Congress (hover/scroll to interact/see details... tools on lower left)")
plt.xlabel("pca 1")
plt.ylabel("pca 2")

#     print("Ron Paul is around (5,2)")
mpld3.display()
```
